```python
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

housing = pd.read_csv('cal_housing_clean.csv')
housing.head()
```

```python
x_train = pd.DataFrame(data=scaler.transform(x_train),columns=x_train.columns,index=x_train.index)
```

```python
x_test = pd.DataFrame(data=scaler.transform(x_test),columns=x_test.columns,index=x_test.index)
```

```python
housing.columns
```

```python
age = tf.feature_column.numeric_column('totalRooms')
rooms = tf.feature_column.numeric_column('totalBedrooms')
pop = tf.feature_column.numeric_column('population')
households = tf.feature_column.numeric_column('households')
income = tf.feature_column.numeric_column('medianIncome')
```

```python
featue_cols = [age,rooms,pop,households,income]
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,batch_size=10,num_epochs=1000,shuffle=True)
test_input_func = tf.estimator.inputs.pandas_input_fn(x=x_test,y=y_test,batch_size=10,num_epochs=1000,shuffle=False)
```

```python
model = tf.estimator.DNNRegressor(hidden_units=[6,10,6],feature_columns=featue_cols)
```

```python
model.train(input_fn=input_func,steps=10000)
```

```python
predict_input_func = tf.estimator.inputs.pandas_input_fn(x_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
pred_gen = model.predict(predict_input_func)
```

```python
prediction = list(pred_gen)
```
